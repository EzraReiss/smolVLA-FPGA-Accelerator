# Downloading Quantized Weights

This jupyter notebook downloads all model quantized onnx weights. This only needs to be run once and is required because we don't commit model weights.



```python
from huggingface_hub import snapshot_download
repo_id = "ainekko/smolvla_base_onnx" # Replace with the actual repository ID
extensions = "*.onnx"  # Specify the file extension to download

local_dir = snapshot_download(repo_id=repo_id, allow_patterns=[extensions], local_dir="./downloads")
print(f"Repository downloaded to: {local_dir}")
```

# Downloading Weights

This jupyter notebook downloads all model weights (450M) for smolVLA model. This only needs to be run once and is required because we don't commit model weights.



## Download Weights

```python
from huggingface_hub import hf_hub_download


repo_id = "lerobot/smolvla_base"  
filename = "./model.safetensors"  

local_file_path = hf_hub_download(
    repo_id=repo_id, 
    filename=filename,
    local_dir="./downloads"
)
```
